```python
import pandas as pd

# Read the Excel file
excel_file = "/content/drive/Shareddrives/FIT5196_S1_2024/A1/Students data/Task 2/Group065.xlsx"

# Initialize an empty list to store filtered DataFrames for each sheet
dfs = []

# Iterate through each sheet in the Excel file
for sheet_name in pd.ExcelFile(excel_file).sheet_names:
    # Read the current sheet into a DataFrame
    df = pd.read_excel(excel_file, sheet_name=sheet_name)

    print(f"DataFrame for Sheet '{sheet_name}':")
    print(df.head())
    print(f"Original DataFrame size for Sheet '{sheet_name}': {df.shape}")

    # Filter columns where 'id' or 'snippet' is present in any row
    filtered_df = df.dropna(axis=1, how='all')

    # Drop rows where both columns are NA
    filtered_df = filtered_df.dropna(subset=filtered_df.columns, how='all')

    # Generate new column names based on position
    new_columns = [f'new_column_{i+1}' for i in range(len(filtered_df.columns))]

    # Rename columns
    filtered_df.columns = new_columns

    # Print the head of the filtered DataFrame
    print(f"Filtered DataFrame for Sheet '{sheet_name}':")
    print(filtered_df.head())

    # Print the size after filtering
    print(f"Filtered DataFrame size for Sheet '{sheet_name}': {filtered_df.shape}")

    # Append the filtered DataFrame to the list
    dfs.append(filtered_df)
```

```python
import ast

def extract_text_original(snippet):
    # Safely evaluate the string representation of the dictionary into a dictionary object
    snippet_dict = ast.literal_eval(snippet)

    # Extracting the 'textOriginal' field from the nested dictionary structure
    if 'topLevelComment' in snippet_dict:
        comment_snippet = snippet_dict['topLevelComment']['snippet']
        return comment_snippet.get('textOriginal', '')
    else:
        return ''

# Create a copy of the DataFrame
text_df = df.copy()

# Apply the function to the 'snippet' column of the copied DataFrame
text_df['textOriginal'] = text_df['snippet'].apply(extract_text_original)

# Now, df_copy contains the 'textOriginal' column extracted from the 'snippet' column
print(text_df.textOriginal.head(50))
```

Since the comment data contain emojis, you will need to remove the emojis and
normalise the text into lower case for further analysis.
 To remove emojis, make sure your text data is in utf-8 format.
 The list of emojis to remove are in emoji.txt.

```python
import re

# Load the list of emojis from emoji.txt
with open('/content/drive/Shareddrives/FIT5196_S1_2024/A1/emoji.txt', 'r', encoding='utf-8') as file:
    emojis = file.read().splitlines()

# Create a regex pattern to match all emojis
emoji_pattern = re.compile('|'.join(re.escape(emoji) for emoji in emojis))

# Function to remove emojis from text
def remove_emojis(text):
    return emoji_pattern.sub('', text)

# Apply the function to the 'textOriginal' column of the DataFrame
text_df['textOriginal'] = text_df['textOriginal'].apply(remove_emojis)

# Normalize the text into lowercase
text_df['textOriginal'] = text_df['textOriginal'].str.lower()

# Print the DataFrame to verify the changes
print(text_df['textOriginal'].head(50))
```

```python
from langdetect import detect, DetectorFactory
from sklearn.feature_extraction.text import CountVectorizer

# Set the random seed for language detection consistency
DetectorFactory.seed = 0

# Function to detect the language of a comment
def detect_language(comment):
    try:
        return detect(comment)
    except:
        return None


# Detect the language of each comment and add it as a new column in the DataFrame
text_df['detected_language'] = text_df['textOriginal'].apply(detect_language)
text_df
```

```python
text_df
```

```python
def extract_channel_id(snippet):
    snippet_dict = ast.literal_eval(snippet)
    if 'channelId' in snippet_dict:
        return snippet_dict['channelId']
    else:
        return None

# Apply the function to extract channelId
text_df['channelId'] = text_df['snippet'].apply(extract_channel_id)
text_df.head(20)
```

```python
# Filter out non-English comments
english_comments_df = text_df[text_df['detected_language'] == 'en']
english_comments_df
```

```python
# Now we have 'channelId' extracted, we can proceed with grouping
english_comments_count = english_comments_df.groupby('channelId').size().reset_index(name='comment_count')
english_comments_count
```

```python
# Filter out channels with at least 15 English comments
channels_with_15_comments = english_comments_count[english_comments_count['comment_count'] >= 15]
channels_with_15_comments.head(366)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer
vectorizer = CountVectorizer()

# Initialize lists to store vocabulary and count vectors for each channel
vocab_lists = []
count_vec_lists = []

# Iterate over each channel with at least 15 English comments
for channel_id in channels_with_15_comments['channelId']:
    # Filter comments for the current channel
    channel_comments = english_comments_df[english_comments_df['channelId'] == channel_id]['textOriginal']

    # Tokenize comments
    tokenized_comments = channel_comments.str.split()

    # Flatten the list of tokenized comments
    all_tokens = [token for sublist in tokenized_comments for token in sublist]

    # Build vocabulary
    vocab = list(set(all_tokens))

    # Generate count vectors
    count_vec = vectorizer.fit_transform(channel_comments)

    # Append vocabulary and count vectors to the lists
    vocab_lists.append(vocab)
    count_vec_lists.append(count_vec)

# Now vocab_lists and count_vec_lists contain the vocabulary and count vectors for each channel respectively
```

```python
# Print vocabulary and count vectors for each channel
for i, channel_id in enumerate(channels_with_15_comments['channelId']):
    print(f"Channel ID: {channel_id}")
    print(f"Vocabulary: {vocab_lists[i]}")
    print(f"Count Vectors: {count_vec_lists[i]}")
    print("="*50)
```
